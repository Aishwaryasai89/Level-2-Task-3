```python
import folium
```

```python
map_center = [df['Latitude'].iloc[0], df['Longitude'].iloc[0]]
restaurant_map =  folium.Map(location=map_center, zoom_start=13)
for index, row in df.iterrows():
    folium. Marker(
        location =  [row['Latitude'], row['Longitude']], 
        popup=row['Restaurant Name'],
        icon=folium.Icon(icon='cutlery', prefix='fa')
        ).add_to(restaurant_map)
restaurant_map.save('restaurant_map.html')
```

```python
from sklearn.cluster import KMeans
```

```python
X = df[['Latitude', 'Longitude']]

num_clusters = 5

kmeans = KMeans(n_clusters=num_clusters, random_state=42)
df['cluster_label'] = kmeans.fit_predict(X)
map_center =  [df['Latitude'].iloc[0], df['Longitude'].iloc[0]]
cluster_map = folium.Map (location=map_center, zoom_start=13)

cluster_colors = ['red', 'blue', 'green', 'purple', 'orange']

for index, row in df.iterrows():
    folium.Marker(
        location = [row['Latitude'], row['Longitude']], 
        popup=row['Restaurant Name'], 
        icon=folium.Icon(color=cluster_colors[row['cluster_label']], icon='cutlery', prefix='fa')
    ).add_to(cluster_map)

cluster_map.save('restaurant_clusters_map.html')
```
